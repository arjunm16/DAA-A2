```python
import matplotlib.pyplot as plt
import numpy as np

f1 = open("op.txt", "r")
lines = f1.readlines()
f1.close()

line_eqns = []
pts_idx = []

for line in lines:
    items = line.rstrip().split()
    print(items)
    leqn = []
    pts_idx.append(int(items[0]))
    leqn.append(float(items[1]))
    leqn.append(float(items[2]))
    line_eqns.append(leqn)

line_eqns.reverse()
pts_idx.reverse()

f1 = open("ip.txt", "r")
lines = f1.readlines()
f1.close()

pts_x = []
pts_y = []

for line in lines[1:]:
    items = line.rstrip().split()
    pts_x.append(float(items[0]))
    pts_y.append(float(items[1]))

plt.figure()
plt.plot(pts_x, pts_y, 'ro')

print(type(pts_idx[0]))

prev = 0
for i in range(0, len(pts_idx)):

    x = np.linspace(prev, pts_x[pts_idx[i]], 100)
    y = line_eqns[i][0] * x + line_eqns[i][1]
    plt.plot(x, y, 'b-')

    # plt.plot(pts_x[i:i+2], pts_y[i:i+2], 'k-')
    prev = pts_x[pts_idx[i]]
    # prev = pts_x[pts_idx[i] + 1]

# plt.show()
```
